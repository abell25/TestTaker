```python
def printQuestions(path, num_questions=None):
    df = pd.read_csv(path, sep='\t')
    num_questions = df.values.shape[0] if num_questions is None else num_questions
    print("Printing {0} / {1} questions\n{2}\n".format(num_questions, df.values.shape[0], '='*60))
    
    Q, A,B,C,D = zip(*df[['question', 'answerA', 'answerB', 'answerC', 'answerD']].head(num_questions).values)
    QnA = zip(Q, A,B,C,D)
    print('\n'.join(['{0}\n\n* {1}\n* {2}\n* {3}\n* {4}\n\n{5}\n'.format(x[0], x[1], x[2], x[3], x[4], '='*60) for x in QnA]))
    #!head -n 10 multipleChoiceQuestions/training_set.tsv
    
printQuestions('multipleChoiceQuestions/validation_set.tsv', 100)
```
